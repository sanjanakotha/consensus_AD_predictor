```python
import protfasta
from itertools import islice

def take(n, iterable):
    """Return the first n items of the iterable as a list."""
    return list(islice(iterable, n))
```

```python
seq_dict = protfasta.read_fasta("../../data/yeast_TFs.fasta")
test_dict = {list(seq_dict.keys())[0]: list(seq_dict.values())[0],
            list(seq_dict.keys())[1]: list(seq_dict.values())[1]}
test_dict
```

```python
protfasta.write_fasta(test_dict, "../../data/test_yeast_TFs_for_tada.fasta")
```
