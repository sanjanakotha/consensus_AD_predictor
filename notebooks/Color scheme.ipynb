```python
from palettable.cubehelix import perceptual_rainbow_16
import matplotlib.pyplot as plt

# Get the list of 16 RGB colors (already normalized)
colors_16 = perceptual_rainbow_16.colors

# Select 8 colors by their indices
indices = [0, 2, 5, 8, 11, 14, 15]
colors_8 = [colors_16[i] for i in indices]

# Display the selected 8-color palette
fig, ax = plt.subplots(figsize=(8, 1))
for i, color in enumerate(colors_8):
    ax.add_patch(plt.Rectangle((i, 0), 1, 1, color=[c/255 for c in color]))
    ax.text(i + 0.5, -0.1, str(color), ha='center', va='top', fontsize='x-small')
ax.set_xlim(0, 8)
ax.set_ylim(0, 1)
ax.axis('off')
plt.show()
```

```python
sns_palette = [tuple(c/255 for c in color) for color in colors_8]
sns.set_palette(sns_palette)
sns.color_palette()
```
